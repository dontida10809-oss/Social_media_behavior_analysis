```python
import pandas as pd

df = pd.read_csv(r"..\data\raw\social_media_addiction.csv")
df.head()
```

```python
df.columns
```

```python
df.shape
df.info()
```

```python
df.isnull().sum()
```

```python
# คอลัมน์ตัวเลข
num_cols = [
    "Age",
    "Avg_Daily_Usage_Hours",
    "Sleep_Hours_Per_Night",
    "Mental_Health_Score",
    "Addicted_Score"
]

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# คอลัมน์ประเภทข้อความ
cat_cols = [
    "Gender",
    "Academic_Level",
    "Country",
    "Most_Used_Platform",
    "Affects_Academic_Performance",
    "Relationship_Status",
    "Conflicts_Over_Social_Media"
]

df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

ค่าตัวเลขแทนด้วยค่าเฉลี่ย ค่าประเภทกลุ่มแทนด้วยค่าที่พบมากที่สุดเพื่อลดผลกระทบต่อการวิเคราะห์เชิงสถิติ

```python
# อายุควรอยู่ในช่วงสมเหตุสมผล
df = df[(df["Age"] >= 15) & (df["Age"] <= 60)]

# เวลาการใช้ Social Media ต่อวัน
df = df[(df["Avg_Daily_Usage_Hours"] >= 0) & (df["Avg_Daily_Usage_Hours"] <= 24)]

# ชั่วโมงการนอน
df = df[(df["Sleep_Hours_Per_Night"] >= 0) & (df["Sleep_Hours_Per_Night"] <= 24)]
```

กรองข้อมูล

```python
df["Usage_Level"] = pd.cut(
    df["Avg_Daily_Usage_Hours"],
    bins=[0, 2, 5, 24],
    labels=["Low", "Medium", "High"]
)
```

ระดับการใช้โซเชียลมีเดีย

```python
df["Addiction_Level"] = pd.cut(
    df["Addicted_Score"],
    bins=[0, 3, 6, 10],
    labels=["Low", "Moderate", "High"]
)
```

ระดับการเสพติดโซเชีบลมีเดีย
